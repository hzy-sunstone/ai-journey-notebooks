```python
import pandas as pd
import numpy as np

# 生成x的取值范围（从-5到5，步长0.1）
x = np.arange(-5, 5.1, 0.1)

# 计算红线（原始模型）的y值
y_red = 1.5 * x**2
# 生成红线的DataFrame并保存为CSV
df_red = pd.DataFrame({'x': x, 'y': y_red})
df_red.to_csv('red_line.csv', index=False)

# 计算蓝线（迁移学习后模型）的y值
y_blue = 1.8 * x**2 + 10
# 生成蓝线的DataFrame并保存为CSV
df_blue = pd.DataFrame({'x': x, 'y': y_blue})
df_blue.to_csv('blue_line.csv', index=False)
```
