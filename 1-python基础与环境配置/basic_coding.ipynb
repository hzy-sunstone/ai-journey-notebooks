```python
import matplotlib.pyplot as plt
import numpy as np

# 生成数据
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# 创建2行1列的第1个子图（上方）
plt.subplot(211)
plt.plot(x, y1, color='blue')
plt.title('正弦曲线')

# 创建2行1列的第2个子图（下方）
plt.subplot(212)
plt.plot(x, y2, color='red')
plt.title('余弦曲线')

# 调整子图间距，避免标题重叠
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. 生成 x 列数据（1到10）
x = list(range(1, 11))  

# 2. 生成 y 列数据（根据 y = 2x + 5）
y = [2 * num + 5 for num in x]  

# 3. 组合成DataFrame
df = pd.DataFrame({
    "x": x,
    "y": y
})

# 4. 保存为CSV文件（无行索引）
df.to_csv("generated_data.csv", index=False)
```

```python
import pandas as pd
import numpy as np

# 设置随机种子（确保结果可复现）
np.random.seed(42)
n_samples = 500  # 样本数量

# 模拟原有特征（收入、房龄、房间数、人口）
avg_area_income = np.random.normal(70000, 15000, n_samples)   # 区域平均收入
avg_area_house_age = np.random.normal(6, 2, n_samples)        # 房屋平均年龄
avg_area_rooms = np.random.normal(5, 1, n_samples)            # 平均房间数
area_population = np.random.normal(35000, 10000, n_samples)   # 区域人口

# 新增特征：房屋面积（Size），假设均值2000平方英尺，标准差300
size = np.random.normal(2000, 300, n_samples)  

# 模拟房价：特征线性组合 + 随机噪声（新增Size的影响：每平方英尺影响50美元）
noise = np.random.normal(0, 10000, n_samples)
price = (
    2 * avg_area_income +           # 收入对房价的贡献
    15000 * avg_area_house_age +    # 房龄对房价的贡献
    10000 * avg_area_rooms +        # 房间数对房价的贡献
    1 * area_population +           # 人口对房价的贡献
    50 * size +                     # 房屋面积对房价的贡献（每平+50美元）
    100000 + noise                  # 基础房价 + 随机波动
)

# 组合成DataFrame（包含Size列）
data = {
    'Avg Area Income': avg_area_income,
    'Avg Area House Age': avg_area_house_age,
    'Avg Area Number of Rooms': avg_area_rooms,
    'Area Population': area_population,
    'Size': size,  # 新增的“房屋面积”列
    'Price': price
}
df = pd.DataFrame(data)

# 保存为CSV文件
df.to_csv('usa_housing_price.csv', index=False)

print("含Size列的模拟房价数据已生成，文件：usa_housing_price.csv")
```
