```python
import numpy as np
import pandas as pd

# --------------------- 生成各区域数据 ---------------------
# 1. 左下区域：ok（X1∈[0,0.5], X2∈[0,0.5]）
n_ok_lower = 120  # 点的数量，可根据需要调整
x1_ok_lower = np.random.uniform(0, 0.5, n_ok_lower)
x2_ok_lower = np.random.uniform(0, 0.5, n_ok_lower)
ok_lower = pd.DataFrame({"X1": x1_ok_lower, "X2": x2_ok_lower, "label": 1})

# 2. 右上区域：ok（X1∈[0.5,1.0], X2∈[0.6,1.0]）
n_ok_upper = 120
x1_ok_upper = np.random.uniform(0.5, 1.0, n_ok_upper)
x2_ok_upper = np.random.uniform(0.5, 1.0, n_ok_upper)
ok_upper = pd.DataFrame({"X1": x1_ok_upper, "X2": x2_ok_upper, "label": 1})

# 3. 左上区域：ng（X1∈[0,0.5], X2∈[0.6,1.0]）
n_ng_upper = 120
x1_ng_upper = np.random.uniform(0, 0.5, n_ng_upper)
x2_ng_upper = np.random.uniform(0.5, 1.0, n_ng_upper)
ng_upper = pd.DataFrame({"X1": x1_ng_upper, "X2": x2_ng_upper, "label": 0})

# 4. 右下区域：ng（X1∈[0.5,1.0], X2∈[0,0.5]）
n_ng_lower = 120
x1_ng_lower = np.random.uniform(0.5, 1.0, n_ng_lower)
x2_ng_lower = np.random.uniform(0, 0.5, n_ng_lower)
ng_lower = pd.DataFrame({"X1": x1_ng_lower, "X2": x2_ng_lower, "label": 0})

# --------------------- 合并并保存 ---------------------
df = pd.concat([ok_lower, ok_upper, ng_upper, ng_lower], ignore_index=True)
df.to_csv("data1.csv", index=False)  # 保存为CSV文件

import pandas as pd
import matplotlib.pyplot as plt

# 1. 读取生成的数据集
df = pd.read_csv("data1.csv")

# 2. 分离不同标签的数据
ok_data = df[df["label"] == 1]   # 筛选 label 为 "ok" 的数据
ng_data = df[df["label"] == 0]   # 筛选 label 为 "ng" 的数据

# 3. 绘制散点图
plt.figure(figsize=(8, 6))  # 设置图的大小
# 绘制 ok 类数据（蓝色散点）
plt.scatter(ok_data["X1"], ok_data["X2"], color="blue", label="ok", alpha=0.7)
# 绘制 ng 类数据（橙色散点）
plt.scatter(ng_data["X1"], ng_data["X2"], color="orange", label="ng", alpha=0.7)

# 4. 图表标注与展示
plt.xlabel("X1")          # X 轴标签
plt.ylabel("X2")          # Y 轴标签
plt.title("Visualization of Simulated Raw Data")  # 图标题
plt.legend()              # 显示图例
plt.grid(alpha=0.3)       # 添加浅灰色网格（增强可读性）
plt.show()                # 显示图像
```
