```python
import pandas as pd
import numpy as np
 
# 加载iris数据集
data=pd.read_csv('iris_dataset.csv')
data.head()
```

```python
x=data.drop(['target','label'],axis=1)
y=data.loc[:,'label']
print(x.shape,y.shape)
```

```python
from sklearn import tree
dc_tree=tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=5)
dc_tree.fit(x,y)
```

```python
y_predict=dc_tree.predict(x)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y,y_predict)
print(accuracy)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
fig=plt.figure(figsize=(20,20))
tree.plot_tree(dc_tree,filled=True,feature_names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'],class_names=['setosa', 'versicolor', 'virginida'])
```
